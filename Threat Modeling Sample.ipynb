```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Threat': ['Data Breach', 'Phishing Attack', 'Malware', 'DDoS Attack', 'Insider Threat'],
    'Likelihood': [0.8, 0.7, 0.6, 0.4, 0.5],  # Scale 0 (low) to 1 (high)
    'Impact': [0.9, 0.5, 0.7, 0.6, 0.8]
}
df = pd.DataFrame(data)
```

```python
# Calculate Risk
df['Risk'] = df['Likelihood'] * df['Impact']
print(df)

# Visualize Da Risk
plt.figure(figsize=(10, 6))
plt.title('Threat Modeling: Risk Matrix')
sns.scatterplot(data=df, x='Likelihood', y='Impact', size='Risk', sizes=(100, 1000), hue='Threat', legend=False, alpha=0.7)
```

```python
for i in range(len(df)):
    plt.text(df['Likelihood'][i] + 0.01, df['Impact'][i] + 0.01, df['Threat'][i], fontsize=12)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('Likelihood')
plt.ylabel('Impact')
plt.grid(True)
plt.show()
```

```python
top_threats = df.sort_values('Risk', ascending=False).head(3)
print("Top 3 Threats by Risk:")
print(top_threats[['Threat', 'Risk']])
```
